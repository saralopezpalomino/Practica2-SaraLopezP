```python
import pandas as pd
import numpy as np
```

## 1. Creación de DataFrames

```python
df1 = pd.DataFrame({"numeros":range(20)})
print("Punto A) DataFrame df1: \n",df1.head(20))
```

```python
df2 = pd.DataFrame ({"aleatorios": np.random.rand(10)})
print("Punto B) DataFrame df2: \n",df2.head(10))
```

```python
df3 = pd.concat([df1, df2], axis=1)
print("Punto C)  DataFrame df3 combinado: \n",df3.head(20))
```

```python
df3.to_csv("df3_combinado.csv", index=False)
print("Punto D) DataFrame df3 guardado en 'df3_combinado.csv'")
```
